# 7. Are there any outliers in the quantity(number of courses completed) attribute?

```python
from scipy.stats import zscore

z_scores = zscore(df['Quantity'])

# Identify outliers based on z-scores
outliers = df[abs(z_scores) > 3]  # You can adjust the threshold value

# Print the potential outliers
print(outliers)
```

```python
# Visualize data distribution using a histogram
plt.figure(figsize=(8, 6))
plt.hist(df['Quantity'], bins=20, edgecolor='k')
plt.xlabel('Quantity (Number of Courses Completed)')
plt.ylabel('Frequency')
plt.title('Distribution of Quantity')
plt.show()

# Calculate z-scores
z_scores = (df['Quantity'] - df['Quantity'].mean()) / df['Quantity'].std()

# Set a threshold for identifying outliers (e.g., z_threshold = 3)
z_threshold = 3

# Identify outliers using z-scores
outliers = df[abs(z_scores) > z_threshold]
# outliers
# print(outliers[['First Name', 'Quantity']])
```

```python
z_scores = (df['Quantity'] - df['Quantity'].mean()) / df['Quantity'].std()

# Define z-score threshold for outliers
z_score_threshold = 3

# Identify outliers using z-scores
potential_outliers = df[abs(z_scores) > z_score_threshold]

# Calculate IQR
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)
IQR = Q3 - Q1

# Define IQR bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers using IQR
iqr_outliers = df[(df['Quantity'] < lower_bound) | (df['Quantity'] > upper_bound)]

# Visualization
plt.figure(figsize=(10, 6))
plt.hist(df['Quantity'], bins=20, edgecolor='k', alpha=0.7)
plt.scatter(potential_outliers.index, potential_outliers['Quantity'], color='r', label='Potential Outliers (Z-Score)')
plt.scatter(iqr_outliers.index, iqr_outliers['Quantity'], color='g', label='Outliers (IQR)')
plt.xlabel('Quantity (Number of Courses Completed)')
plt.ylabel('Frequency')
plt.legend()
plt.title('Distribution of Quantity Attribute')
plt.show()
```

## 13. Do students in leadership positions during their college years tend to have higher GPAs or better expected salary?

```python
from scipy.stats import ttest_ind


# Define a leadership indicator column (0 for no leadership, 1 for leadership)
df['leadership_indicator'] = (df['Leadership- skills'] == 'Yes').astype(int)

# Data analysis
average_cgpa_leadership = df.groupby('leadership_indicator')['CGPA'].mean()
average_salary_leadership = df.groupby('leadership_indicator')['Expected salary (Lac)'].mean()

# Statistical testing (t-test)
cgpa_leadership = df[df['leadership_indicator'] == 1]['CGPA']
cgpa_no_leadership = df[df['leadership_indicator'] == 0]['CGPA']
t_statistic_cgpa, p_value_cgpa = ttest_ind(cgpa_leadership, cgpa_no_leadership)

# Similar testing for expected salary
salary_leadership = df[df['leadership_indicator'] == 1]['Expected salary (Lac)']
salary_no_leadership = df[df['leadership_indicator'] == 0]['Expected salary (Lac)']
t_statistic_salary, p_value_salary = ttest_ind(salary_leadership, salary_no_leadership)

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=average_cgpa_leadership.index, y=average_cgpa_leadership.values)
plt.xlabel('Leadership Indicator')
plt.ylabel('Average CGPA')
plt.title('Average CGPA Comparison: Leadership vs. No Leadership')
plt.xticks([0, 1], ['No Leadership', 'Leadership'])
plt.show()

# Visualization for expected salary
plt.figure(figsize=(10, 6))
sns.barplot(x=average_salary_leadership.index, y=average_salary_leadership.values)
plt.xlabel('Leadership Indicator')
plt.ylabel('Average Expected Salary (Lac)')
plt.title('Average Expected Salary Comparison: Leadership vs. No Leadership')
plt.xticks([0, 1], ['No Leadership', 'Leadership'])
plt.show()

# Print t-test results
print("CGPA T-Test Results:")
print(f"T-statistic (CGPA): {t_statistic_cgpa}")
print(f"P-value (CGPA): {p_value_cgpa}")

print("Expected Salary T-Test Results:")
print(f"T-statistic (Expected Salary): {t_statistic_salary}")
print(f"P-value (Expected Salary): {p_value_salary}")
```

## 15. Which promotion channel brings in more student participations for the event ?

```python
# Preprocess the data: split multiple options and create a list of promotion channels
df['promotion_channels'] = df['How did you come to know about this event?'].apply(
    lambda x: [] if pd.isna(x) else x.split(', ')
)

# Flatten the list of promotion channels
flattened_channels = [channel for channels_list in df['promotion_channels'] for channel in channels_list]

# finding the most common promotion channel
most_common_promotion_channel = Counter(flattened_channels).most_common(1)[0]

total_participants = len(df)

print("Most Common Promotion Channel:", most_common_promotion_channel[0])
print(f"Number of Participants from Most Common Channel '{most_common_promotion_channel[0]}': {most_common_promotion_channel[1]}")
```

#### Conclusion:
By preprocessing the data to account for multiple promotion channels and then counting the occurrences of each channel, the code identifies the promotion channel that brings in the most student participation. The conclusion is the determination of the most common promotion channel and the corresponding count of participants associated with that channel.

## 17. Those who have high CGPA & More experience in language those who had high expectations for salary? (Avg)

```python
import math
df['Experience with python (Months)'] = pd.to_numeric(df['Experience with python (Months)'], errors='coerce')

# Calculate the median CGPA and median experience with Python
median_cgpa = df['CGPA'].median()
median_experience = df['Experience with python (Months)'].median()

# Define thresholds based on medians (you can adjust these values as needed)
high_cgpa_threshold = median_cgpa
high_experience_threshold = median_experience

# Filtering students with high CGPA and more experience
filtered_students = df[(df['CGPA'] >= high_cgpa_threshold) & (df['Experience with python (Months)'] >= high_experience_threshold)]

# Calculate the average expected salary for the filtered group
average_salary_high_cgpa_experience = filtered_students['Expected salary (Lac)'].mean()
average_salary_rounded = math.ceil(average_salary_high_cgpa_experience * 100) / 100

print(f"Average Expected Salary for Students with High CGPA and More Experience: {average_salary_rounded} Lakh")
```
